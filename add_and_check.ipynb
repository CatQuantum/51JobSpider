```python
import pandas as pd

import os
# 读取Excel文件A
df_A = pd.read_excel(r'C:\Users\Xue\Documents\GitHub\gender-discrimination-hiring\Data_IT\所有IT类别.xlsx', engine='openpyxl')


# 设置包含Excel文件的文件夹路径
folder_path = r'C:\Users\Xue\Documents\GitHub\gender-discrimination-hiring\Data_IT\new\applied'

# 创建一个空的DataFrame用于合并数据
merged_df = pd.DataFrame()

# 遍历指定文件夹中的所有文件
for filename in os.listdir(folder_path):
    # 检查文件扩展名是否为.xlsx，确保只处理Excel文件
    if filename.endswith('.xlsx'):
        # 构造完整的文件路径
        file_path = os.path.join(folder_path, filename)
        # 读取Excel文件
        df = pd.read_excel(file_path, engine='openpyxl')
        # 将读取的数据合并到总的DataFrame中
        merged_df = pd.concat([merged_df, df], ignore_index=True)

# 合并B类文件的DataFrame
df_B = merged_df

# 使用"company_name"作为唯一标识符
# 去重，确保company_name的唯一性
df_B = df_B.drop_duplicates(subset=['Company'])

# 找出只存在于A的公司名称
only_in_A = df_A[~df_A['fullCompanyName'].isin(df_B['Company'])]

# 找出只存在于B的公司名称
only_in_B = df_B[~df_B['Company'].isin(df_A['fullCompanyName'])]

# 找出两者共有的公司名称
in_both = df_A[df_A['fullCompanyName'].isin(df_B['Company'])]

# 输出结果
print("只存在于A的公司有：")
print(only_in_A)

print("只存在于B的公司有：")
print(only_in_B)

# print("两者共有的公司有：")
# print(in_both)
```
